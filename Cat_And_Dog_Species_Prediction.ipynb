```python
class Model(nn.Module):

  def __init__(self):
    super(Model, self).__init__()
    pretrained_model = resnet18(weights=ResNet18_Weights.DEFAULT)
    self.backbone = nn.Sequential(*list(pretrained_model.children())[:-2])

    self.resnet_output_shape = 512 * 7 * 7      # for each image

    self.fc1 = nn.Linear(self.resnet_output_shape, 1024)
    self.fc2 = nn.Linear(1024, 256)
    self.fc3 = nn.Linear(256, 64)

    self.cat_dog_or_none_layer = nn.Linear(64, 3)
    self.species_layer = nn.Linear(64 + 3, 10)
    self.bounding_boxes = nn.Linear(64 + 3 + 10, 4)


  def forward(self, images):
    features = self.backbone(images)
    flattened_features = features.view(features.size(0), -1)

    features = self.fc1(flattened_features)
    features = F.leaky_relu(features, negative_slope=0.2)
    features = self.fc2(features)
    features = F.leaky_relu(features, negative_slope=0.2)
    features = self.fc3(features)
    features = F.leaky_relu(features, negative_slope=0.2)

    cat_dog_or_none = self.cat_dog_or_none_layer(features)
    cat_dog_or_none_softmax = F.softmax(cat_dog_or_none, dim=1)

    species = self.species_layer(torch.cat([features, cat_dog_or_none_softmax], dim=1))
    species_softmax = F.softmax(species, dim=1)

    bounding_boxes = self.bounding_boxes(torch.cat([features, cat_dog_or_none_softmax, species_softmax], dim=1))

    return {"cat_dog_or_none": cat_dog_or_none_softmax, "species": species_softmax, "bounding_box" : bounding_boxes}
```

```python
from bs4 import BeautifulSoup

def read_xml_file(path):
    with open(path, 'r') as f:
        data = f.read()

    data = BeautifulSoup(data, 'xml')

    cat_dog_or_none = 2     # instead of checking if object exists I am using 3 classes, cat dog or non

    if data.find('name') and (data.find('name').text == 'dog'):
        cat_dog_or_none = 1

    elif data.find('name') and (data.find('name').text == 'cat'):
        cat_dog_or_none = 0

    xmin = 0
    ymin = 0
    xmax = 0
    ymax = 0

    bounding_box = [xmin, ymin, xmax, ymax]

    species_index = 9     # 9 is for none

    if not(cat_dog_or_none == 2):
      xmin = int(data.find("xmin").text)
      ymin = int(data.find("ymin").text)
      xmax = int(data.find("xmax").text)
      ymax = int(data.find("ymax").text)
      
      bounding_box = [xmin, ymin, xmax, ymax]
      
      species = "_".join(path.split(os.sep)[-1].split("_")[:-1])

      if species in species_to_index:
            species_index = species_to_index[species]
      else:
          species_index = len(species_to_index)
          species_to_index[species] = species_index
          index_to_species[species_index] = species

    features = {"cat_dog_or_none": cat_dog_or_none, "species": species_index, "bounding_box" : torch.tensor(bounding_box)}

    return features
```

```python
class CustomDataset(torch.utils.data.Dataset):
    def __init__(self, dataset_path, images_list):
        self.data_source = dataset_path
        self.images_list = images_list

        image_folder_path = os.path.join(dataset_path, "images")
        label_folder_path = os.path.join(dataset_path, "labels")

        self.preprocess = transforms.Compose([
            transforms.Resize((224, 224)),
            transforms.RandomHorizontalFlip(),
            transforms.ToTensor()
        ])

        self.data_points = []

        for path in os.listdir(image_folder_path):
            name = path.split(os.sep)[-1].split(".")[0]
            extension = path.split(os.sep)[-1].split(".")[1]

            if 'jpg' not in extension:
              continue

            if name in images_list:

              image_path = os.path.join(image_folder_path, f"{name}.jpg")
              image = Image.open(image_path).convert("RGB")
              image = self.preprocess(image)

              xml_path = os.path.join(label_folder_path, f"{name}.xml")
              xml_data = read_xml_file(xml_path)

              self.data_points.append([image, xml_data])

    def __len__(self):
        return len(self.data_points)

    def __getitem__(self, index):
        image, labels = self.data_points[index]
        return image, labels
```

# TRAINING LOOP IMPLEMENTATION

## Custom Loss function


```python
def dynamically_scale(entropy_loss, bounding_loss):
    if bounding_loss > 10000:
      bounding_loss_scaled = bounding_loss/10000

    else:
      bounding_loss_scaled = bounding_loss/1000

    return entropy_loss, bounding_loss_scaled

def compute_loss(predictions, targets, val = False, plot = True):

    cat_dog_or_none_loss = F.cross_entropy(predictions["cat_dog_or_none"], targets["cat_dog_or_none"])
    species_loss = F.cross_entropy(predictions['species'], targets['species'])
    
    entropy_loss = cat_dog_or_none_loss + species_loss
    bounding_box_loss = nn.MSELoss()(predictions['bounding_box'], targets['bounding_box'].float())
    
    entropy_loss, bounding_box_loss_scaled = dynamically_scale(entropy_loss, bounding_box_loss)

    total_loss = entropy_loss + bounding_box_loss_scaled

    if plot:
      entropy_str = "Entropy Loss"
      bounding_str = "Bounding Loss "
      total_str = "Total loss "

      if val:
        entropy_str += ' (val) '
        bounding_str += ' (val) '
        total_str += ' (val) '

      wandb.log({entropy_str: entropy_loss.item()})
      wandb.log({bounding_str: bounding_box_loss.item()})
      wandb.log({total_str: total_loss.item()})

    return total_loss
```

## Custom metric function

```python
def compute_metrics(predictions, targets, val=False, plot = True):
    cat_dog_or_none_pred = torch.argmax(predictions["cat_dog_or_none"], dim=1)
    species_pred = torch.argmax(predictions["species"], dim=1)
    
    cat_dog_or_none_target = targets["cat_dog_or_none"]
    species_target = targets["species"]
    
    cat_dog_or_none_accuracy = accuracy_score(cat_dog_or_none_target.detach().cpu().numpy(), cat_dog_or_none_pred.detach().cpu().numpy())
    species_accuracy = accuracy_score(species_target.detach().cpu().numpy(), species_pred.detach().cpu().numpy())

    cat_dog_or_none_precision = precision_score(cat_dog_or_none_target.detach().cpu().numpy(), cat_dog_or_none_pred.detach().cpu().numpy(), average='weighted', zero_division=1.0)
    species_precision = precision_score(species_target.detach().cpu().numpy(), species_pred.detach().cpu().numpy(), average='weighted', zero_division=1.0)

    cat_dog_or_none_recall = recall_score(cat_dog_or_none_target.detach().cpu().numpy(), cat_dog_or_none_pred.detach().cpu().numpy(), average='weighted', zero_division=1.0)
    species_recall = recall_score(species_target.detach().cpu().numpy(), species_pred.detach().cpu().numpy(), average='weighted', zero_division=1.0)

    bounding_box_mse = mean_squared_error(predictions['bounding_box'].detach().cpu().numpy(), targets['bounding_box'].detach().cpu().numpy())
    
    if plot:
      accuracy_str_cd = "Cat/Dog/None Accuracy"
      accuracy_str_species = "Species Accuracy"
      precision_str_cd = "Cat/Dog/None Precision"
      precision_str_species = "Species Precision"
      recall_str_cd = "Cat/Dog/None Recall"
      recall_str_species = "Species Recall"
      mse_str = "Bounding Box MSE"

      if val:
          accuracy_str_cd += " (val)"
          accuracy_str_species += " (val)"
          precision_str_cd += " (val)"
          precision_str_species += " (val)"
          recall_str_cd += " (val)"
          recall_str_species += " (val)"
          mse_str += " (val)"

      wandb.log({accuracy_str_cd: cat_dog_or_none_accuracy})
      wandb.log({accuracy_str_species: species_accuracy})
      wandb.log({precision_str_cd: cat_dog_or_none_precision})
      wandb.log({precision_str_species: species_precision})
      wandb.log({recall_str_cd: cat_dog_or_none_recall})
      wandb.log({recall_str_species: species_recall})
      wandb.log({mse_str: bounding_box_mse})

    return cat_dog_or_none_accuracy, species_accuracy, cat_dog_or_none_precision, species_precision, cat_dog_or_none_recall, species_recall, bounding_box_mse

def compute_classification_report(predictions, targets):

    cat_dog_or_none_pred = torch.argmax(predictions["cat_dog_or_none"], dim=1)
    species_pred = torch.argmax(predictions["species"], dim=1)
    
    cat_dog_or_none_target = targets["cat_dog_or_none"]
    species_target = targets["species"]

    cat_dog_or_none_report = classification_report(cat_dog_or_none_target.detach().cpu().numpy(), cat_dog_or_none_pred.detach().cpu().numpy())
    species_report = classification_report(species_target.detach().cpu().numpy(), species_pred.detach().cpu().numpy())

    print("Cat/Dog/None Classification Report:")
    print(cat_dog_or_none_report)

    print()
    print("Species Classification Report:")
    print(species_report)
```

## Custom train function


```python
def train(model, optimizer, data_loader, num_epochs, val_loader = None, resume_weights_path = None):

  best_val_loss = float('inf')

  if resume_weights_path is not None:
    model.load_state_dict(torch.load(resume_weights_path))
    model.train()
    print(f"Resuming training from {resume_weights_path}")

  for epoch in tqdm(range(1, num_epochs+1), desc="Epochs"):
      
      for _, data in enumerate(data_loader):
          inputs, labels = data
          inputs = inputs.to(device)
          labels = {k: v.to(device) if torch.is_tensor(v) else v for k, v in labels.items()}

          optimizer.zero_grad()

          output = model(inputs)
          loss = compute_loss(output, labels)
          loss_val = loss.item()

          loss.backward()
          optimizer.step()

      if val_loader:
          for _, data in enumerate(val_loader):
            inputs, labels = data
            inputs = inputs.to(device)
            labels = {k: v.to(device) if torch.is_tensor(v) else v for k, v in labels.items()}

            output = model(inputs)
            val_loss = compute_loss(output, labels, val = True)
            metrics = compute_metrics(output, labels, val = True)

            if val_loss < best_val_loss:
                best_val_loss = val_loss

                if resume_weights_path is not None:
                  torch.save(model.state_dict(), resume_weights_path)
                  print(f"Saved best weights at {resume_weights_path}")
                
                else:
                  torch.save(model.state_dict(), "model_weights.pth")
```

## Custom Test Function

```python
def test(model, test_loader):
    model.eval()
    test_loss = 0.0
    predictions = {"cat_dog_or_none": [], "species": [], "bounding_box": []}
    targets = {"cat_dog_or_none": [], "species": [], "bounding_box": []}

    with torch.no_grad():
        for _, data in enumerate(test_loader):
            inputs, labels = data
            inputs = inputs.to(device)
            labels = {k: v.to(device) if torch.is_tensor(v) else v for k, v in labels.items()}

            output = model(inputs)
            loss = compute_loss(output, labels, val=True, plot = False)
            test_loss += loss.item()

            for key, value in output.items():
                predictions[key].append(value)
            for key, value in labels.items():
                targets[key].append(value)

    test_loss /= len(test_loader)

    predictions = {k: torch.cat(v, dim=0) for k, v in predictions.items()}
    targets = {k: torch.cat(v, dim=0) for k, v in targets.items()}
    metrics = compute_metrics(predictions, targets, val=True, plot = False)
    compute_classification_report(predictions, targets)

    return test_loss, metrics
```

## Custom Visualization Functions

```python
def visualize_bounding_box(image, bbox, heading):
  
    fig, ax = plt.subplots(1)
    ax.imshow(image)
    ax.set_title(heading)

    xmin, ymin, xmax, ymax = bbox
    width = xmax - xmin
    height = ymax - ymin

    rect = patches.Rectangle((xmin, ymin), width, height, linewidth=2, edgecolor='r', facecolor='none')
    ax.add_patch(rect)

    plt.show()

def convert_output_to_standard(output):
    output_bounding_box = output['bounding_box'].detach().cpu().numpy()

    cat_dog_or_none = int(torch.argmax(output["cat_dog_or_none"]).detach().cpu())
    specie = int(torch.argmax(output["species"]).detach().cpu())

    bounding_box = output_bounding_box[0]

    has_object = False
    cat_or_dog = None
    species = None

    if cat_dog_or_none == 2:
      has_object = True

    else:
      cat_or_dog = number_to_class[cat_dog_or_none]

    species = index_to_species[specie]

    return {
              "has_object" : has_object,
              "cat_or_dog" : cat_or_dog,
              "species" : species,
              "xmin" : bounding_box[0],
              "ymin" : bounding_box[1],
              "xmax" : bounding_box[2],
              "ymax" : bounding_box[3]
          }

def convert_target_to_standard(target):
    bounding_box = target['bounding_box'].detach().cpu().numpy()

    cat_dog_or_none = int(target["cat_dog_or_none"])
    specie = int(target["species"])

    has_object = False
    cat_or_dog = None
    species = None

    if cat_dog_or_none == 2:
      has_object = True

    else:
      cat_or_dog = number_to_class[cat_dog_or_none]

    species = index_to_species[specie]

    return {
              "has_object" : has_object,
              "cat_or_dog" : cat_or_dog,
              "species" : species,
              "xmin" : bounding_box[0],
              "ymin" : bounding_box[1],
              "xmax" : bounding_box[2],
              "ymax" : bounding_box[3]
          }

def visualize(input_sample, model, target_sample = None):
    model.eval()

    with torch.no_grad():
      input_tensor = torch.unsqueeze(input_sample, 0)
      input_tensor = input_tensor.to(device)
      output = model(input_tensor)

      output_bounding_box = output['bounding_box'].detach().cpu().numpy()
      numpy_image = input_sample.detach().cpu().numpy()

      if numpy_image.shape[0] == 3:
        numpy_image = numpy_image.transpose(1, 2, 0)

      visualize_bounding_box(numpy_image, output_bounding_box[0], "Obtained bounding box")
      standard_form = convert_output_to_standard(output)

      if target_sample:

        target_bounding_box = target_sample['bounding_box'].detach().cpu().numpy()
        visualize_bounding_box(numpy_image, target_bounding_box, "Target bounding box")
        target_standard = convert_target_to_standard(target_sample)

        return standard_form, target_standard
      
      else:

        return standard_form

```
